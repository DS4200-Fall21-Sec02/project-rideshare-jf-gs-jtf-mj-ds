# Data manipulation

```python
import pandas as pd
df_data = pd.read_csv('all_vis.csv')
df_data.head()
```

```python
type(df_data['datetime'][0])
```

```python
from datetime import datetime
import pandas as pd
import math

def change_date(row):
    ''' reformats a wrds date into a year
    
    Args:
        row (pd.Series): row with date to reformat
        
    Returns:
        row (pd.Series): row with updated date
    '''
    
    # convert the date type
    str_date = str(row.loc['datetime'])
    dt_date = datetime.strptime(str_date, '%Y-%m-%d %H:%M:%S')
    
    weekday = datetime.strftime(dt_date, '%a')
        
    # readjust the series value
    row.loc['weekday'] = weekday
    
    return row
```

```python
df_data = df_data.apply(change_date, axis = 1)
```

```python
df_data.head()
```

```python
import numpy as np
days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']

for day in days:
    
    temp = df_data[df_data['weekday'] == day]
    
    print(day, ' : ', np.mean(temp['surge_multiplier']))
   
```

```python
for day in days:
    
    temp = df_data[df_data['weekday'] == day]
    
    print(day, ' : ', np.max(temp['surge_multiplier']))
```
